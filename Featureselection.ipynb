```python
import os
from pathlib import Path
import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # Added for plotting
from sklearn.model_selection import cross_val_predict, KFold  # Changed to KFold for regression
from sklearn.ensemble import RandomForestRegressor  # Changed to regression model
from sklearn.metrics import mean_squared_error  # Added for MSE calculation
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier  # or whichever model you're using


#sys.path.append(str(Path(__file__).absolute().parents[1]))
 
#os.system('cls' if os.name == 'nt' else 'clear')

LAST_INDEX_TRAIN = -1  # Tudo = -1
LAST_INDEX_TEST = -1  # Tudo = -1


#These columns contain non-feature data (e.g., patient name, age, BMI, DXA scores).
#Dropping them leaves you with only the radiomic features.

COLUMNS_TO_REMOVE = ['Pacientes','Nome','Diagnosticounificado', 'T-Score L1-L4',
                     'EspessuraCorticalPan', 'EspessuraCorticalTrans', 'IMC', 'T-Score Fêmur Total',
                    'Idade', 'Altura', 'Peso', 'BMD L1-L4', 'BMD Colo Femoral', 'BMD Fêmur Total']

TARGET_NAME = 'T-Score Colo Femoral'
#You run separate models on different subgroups of anatomical regions:
SUBGRUPOS = ['_CD', '_CE', '_MD', '_ME', '_SA', 'all']

# 'Diagnosticounificado', 'T-Score L1-L4', 'T-Score Colo Femoral', 'T-Score Fêmur Total'

# Define the path of  CSVs for training and testing
FILE_PATH = 'train_database_matheus.csv'

# Define where the trained model is saved
PATH_TRAINED_MODEL = '/d01/scholles/gigasistemica/matheus-giga/model/radiomics_rf_model.sav'


def read_csv_and_split(path, columns_to_remove):
    # Read the CSV
    df = pd.read_csv(path, sep=';')
    # Remove the specified columns
    df_train = df.drop(columns=columns_to_remove)

    # Get the list of remaining columns
    all_columns = list(df_train.columns)

    return df_train, all_columns

def filtrar_palavras(lista, parametro):
    palavras_especificas = [
        'Altura da cortical (D)','Altura forame-base (D)',
        'Altura da medular forame-superficie  da cortical (D)','Altura da cortical (E)',
        'Altura forame-base (E)','Altura da medular forame-superficie  da cortical (E)',
    ]

    if parametro.lower() == "all":
        return lista
    else:
        palavras_filtradas = [palavra for palavra in lista if parametro.lower() in palavra.lower() or palavra in palavras_especificas]
        return palavras_filtradas
        #“Use subgroup-specific features plus a few trusted VIP features we know are useful across the board.”






if __name__ == '__main__':
    for subgrupo in SUBGRUPOS:
        print(f"\n[INFO] Processing subgroup: {subgrupo}")
        
        # Load and clean data
        train_df, all_columns = read_csv_and_split(FILE_PATH, COLUMNS_TO_REMOVE)
        target = [TARGET_NAME]
        features = filtrar_palavras(all_columns, subgrupo)

        # Remove target column from features (if present)
        if TARGET_NAME in features:
            features.remove(TARGET_NAME)

        X = train_df[features].to_numpy()
        y = train_df[target].to_numpy().ravel()  # Flatten target to 1D

        # Define the model
        model = RandomForestRegressor(verbose=True, random_state=42)

        # === Recursive Feature Elimination (RFE) ===
        num_features_to_select = min(15, X.shape[1])  # Choose top 15 features or fewer if less
        rfe = RFE(estimator=model, n_features_to_select=num_features_to_select, step=1)
        rfe.fit(X, y)

        # Filter features selected by RFE
        selected_feature_indices = np.where(rfe.support_)[0]
        selected_features = [features[i] for i in selected_feature_indices]

        print(f"[INFO] Selected features for {subgrupo} using RFE:")
        for f in selected_features:
            print(" -", f)

        # Redefine X using only selected features
        X_selected = train_df[selected_features].to_numpy()

        # Cross-validation and predictions
        cv = KFold(n_splits=5, shuffle=True, random_state=42)
        y_pred_cv = cross_val_predict(model, X_selected, y, cv=cv)

        # Calculate Mean Squared Error
        mse = mean_squared_error(y, y_pred_cv)
        print(f"Mean Squared Error (MSE) for {subgrupo}: {mse:.4f}")

        # Plot predicted vs true
        plt.figure(figsize=(6, 6))
        plt.scatter(y, y_pred_cv, color='blue', alpha=0.5)
        plt.plot([min(y), max(y)], [min(y), max(y)], 'r--', linewidth=2)
        plt.text(min(y), max(y), f'SUBGRUPO: {subgrupo}', verticalalignment='bottom', horizontalalignment='left', color='green', fontsize=10)
        plt.xlabel('True Values')
        plt.ylabel('Predictions')
        plt.title(f'{TARGET_NAME} - MSE: {mse:.4f}')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```
