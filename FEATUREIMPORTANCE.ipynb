```python
import os
import pickle
import sys
from pathlib import Path
import time
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Define the size of the dataset
LAST_INDEX_TRAIN = -1
LAST_INDEX_TEST = -1

COLUMNS_TO_REMOVE = ['Pacientes','Nome', 'T-Score L1-L4', 'T-Score F√™mur Total',
                     'EspessuraCorticalPan', 'EspessuraCorticalTrans', 'IMC', 'T-Score Colo Femoral',
                    'Idade', 'Altura', 'Peso', 'BMD L1-L4', 'BMD Colo Femoral', 'BMD F√™mur Total']

TARGET_NAME = 'Diagnosticounificado'
SUBGRUPOS = ['all','_CD', '_CE', '_MD', '_ME', '_SA']

FILE_PATH = 'train_database_matheus.csv'

# Output directory to save plots
OUTPUT_DIR = './feature_importance'
os.makedirs(OUTPUT_DIR, exist_ok=True)

def read_csv_and_split(path, columns_to_remove):
    df = pd.read_csv(path, sep=';')
    df_train = df.drop(columns=columns_to_remove)
    all_columns = list(df_train.columns)
    return df_train, all_columns

def filtrar_palavras(lista, parametro):
    palavras_especificas = [
        'Altura da cortical (D)','Altura forame-base (D)',
        'Altura da medular forame-superficie  da cortical (D)','Altura da cortical (E)',
        'Altura forame-base (E)','Altura da medular forame-superficie  da cortical (E)',
    ]
    if parametro.lower() == "all":
        return lista
    else:
        return [palavra for palavra in lista if parametro.lower() in palavra.lower() or palavra in palavras_especificas]

if __name__ == '__main__':
    for subgrupo in SUBGRUPOS:
        train_df, all_columns = read_csv_and_split(FILE_PATH, COLUMNS_TO_REMOVE)
        target = [TARGET_NAME]

        features = filtrar_palavras(all_columns, subgrupo)
        try:
            features.remove(TARGET_NAME)
        except:
            pass

        X = train_df[features].to_numpy()
        y = train_df[target].to_numpy()

        # Model and cross-validation
        learning_model = RandomForestClassifier(verbose=True)
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        y_pred_cv = cross_val_predict(learning_model, X, y.ravel(), cv=cv)

        print(f"\n Classification Report for subgroup: {subgrupo}")
        print(classification_report(y, y_pred_cv, digits=4))

        # Train full model
        print('üõ†Ô∏è Initiating the training...')
        learning_model.fit(X, y.ravel())

        start_time = time.time()
        importances = learning_model.feature_importances_
        std = np.std([tree.feature_importances_ for tree in learning_model.estimators_], axis=0)
        elapsed_time = time.time() - start_time

        print(f"‚è±Ô∏è Elapsed time to compute importances: {elapsed_time:.3f} seconds")

        # Show feature importance table
        forest_importances = pd.Series(importances, index=features)
        feature_importance_table = pd.DataFrame({
            'Feature': forest_importances.index,
            'Importance': forest_importances.values
        }).sort_values(by='Importance', ascending=False)

        print("\n Top 10 Feature Importances:")
        print(feature_importance_table.head(10).to_string(index=False))

        # Plot and save bar chart
        plt.figure(figsize=(16, 12))
        fig, ax = plt.subplots(figsize=(16, 12))
        forest_importances.plot.bar(yerr=std, ax=ax)
        ax.set_title(f"Feature Importance usando MDI - {subgrupo}", fontsize=20)
        ax.set_ylabel("Mean decrease in impurity", fontsize=18)
        plt.tight_layout(pad=4.0)
        plt.show()

        # Save the figure
        fig_path = os.path.join(OUTPUT_DIR, f"feat_importance_{subgrupo}.png")
        fig.savefig(fig_path, dpi=300)
        plt.close()
        print(f" Saved plot to: {fig_path}")
```
